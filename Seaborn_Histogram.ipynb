```python
import seaborn as sns
```

```python
data = sns.load_dataset("penguins") 
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
sns.histplot(x="bill_length_mm" ,data=data)
```

```python
sns.histplot(x="bill_length_mm" ,data=data, kde=True)
```

# Y axis : Count-Density-Probability

```python
sns.histplot(x="bill_length_mm" ,data=data, kde=True, bins=20)
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='density')
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='probability')
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='probability', cumulative=True)
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='probability', cumulative=True, element='step', fill=False)
```

# Add Categorical data

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='count', hue="species")
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='count', hue="species", element="step")
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='count', hue="species", element="poly")
```

```python
sns.histplot(x="bill_length_mm" ,data=data, stat='count', hue="species", multiple="stack")
```

# Bivariant Histogram

```python
sns.histplot(x="bill_length_mm" ,y="species",data=data, stat='count', hue="species")
```
